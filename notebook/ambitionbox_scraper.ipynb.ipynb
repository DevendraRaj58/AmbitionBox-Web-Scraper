```python
#Import reuireed library
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# headers help us disguise ourself as web browser sending requests
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
#below code extracts the entire content on page 1 of the given url, converts into text and stores in variable
response=requests.get('https://www.ambitionbox.com/list-of-companies?page=1',headers=headers).text
```

```python
#Creating beautifulsoup object
soup=BeautifulSoup(response,'lxml')
```

```python
#find_all function helps find the all section where the mentioned tag is found in the given url
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip()) # Strip is used to remove special character in h2 tagged elements...
```

```python
for i in soup.find_all('div', class_='rating_text'): # Scraping rating of each company from first page
    print(i.text.strip())
```

```python
company=soup.find_all('div',class_='companyCardWrapper') # Iterate each wrapper or card of each company
```

```python
#Create separate list for each content scraped which would act as unique column in final dataframe
name=[]
rating=[]
reviews=[]
salary=[]
for i in company :
    name.append(i.find('h2').text.strip())
    rating.append(i.find('div',class_='rating_text').text.strip())
    reviews.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[0].text.strip())
    salary.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[1].text.strip())
    '''Since for the site ambtionbox 4 elements have same tag and class
    we fetch them as per list item'''  


df_dict={'name':name,'rating':rating,'reviews':reviews,'salary':salary}
df=pd.DataFrame(df_dict)
```

```python
# Similarly to iterate thorugh all pages we combine entire code over a loop and dynamically change pages on basis of iterable
final=pd.DataFrame()
for j in range(1,11):
    url='https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    
    response=requests.get(url,headers=headers).text

    soup=BeautifulSoup(response,'lxml')

    company=soup.find_all('div',class_='companyCardWrapper')

    name=[]
    rating=[]
    reviews=[]
    salary=[]
    for i in company :
        name.append(i.find('h2').text.strip())
        rating.append(i.find('div',class_='rating_text').text.strip())
        reviews.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[0].text.strip())
        salary.append(i.find_all('a',class_='companyCardWrapper__ActionWrapper')[1].text.strip())
        '''Since for the site ambtionbox 4 elements have same tag and class
        we fetch them as per list item'''  
    
    
    df_dict={'name':name,'rating':rating,'reviews':reviews,'salary':salary}
    df=pd.DataFrame(df_dict)

    final = pd.concat([final, df], ignore_index=True)
```
